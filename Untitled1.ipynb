```python
def get_s_spans(df,threshold = 0):
    spans = []
    curr = 0
    span = []
    for i in range(len(df)-1):
        curr += 1
        if df.iloc[i,8] == 'True':
            span.append(curr)
            curr = 0
        if df.iloc[i,0] != df.iloc[i+1,0]:
            if curr != 0:
                span.append(curr)
                curr = 0
            if len(span) > threshold:
                spans.append(span)
            span = []
    return spans
            
```

```python
import numpy as np
def memory(s_spans,threshold = 0):
    res = []
    for span in s_spans:
        temp = []
        #print(span)
        m1,s1 = pd.Series(span[:-1]).mean(),pd.Series(span[:-1]).std()
        m2,s2 = pd.Series(span[1:]).mean(),pd.Series(span[1:]).std()
        if  np.isnan(s1) or np.isnan(s2):
            res.append(0)
            #print("LL")
            continue
        #print(m1,s1,m2,s2)
        for i in range(len(span)-1):
            m = ((span[i] - m1)*(span[i+1] - m2))/(s1*s2)
        res.append(m)
    return res
```
